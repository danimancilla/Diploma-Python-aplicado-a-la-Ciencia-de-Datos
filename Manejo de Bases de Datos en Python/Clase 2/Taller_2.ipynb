```python
#Los nombres de las aerolíneas mexicanas ordenadas afabéticamente.
%%sql
select nombre
from aerolinea
where pais = 'Mexico'
order by nombre asc
```

```python
#Los códigos IATA de las aerolíneas que ofrecen vuelos desde Santiago. El código IATA del
#aeropuerto de Santiago es SCL.
%%sql
select distinct aerolinea
from ruta
where origen = 'SCL'
```

```python
#Los nombres de los aeropuertos de la ciudad de New York.
%%sql
select nombre
from aeropuerto
where ciudad = 'New York'
```

```python
#Los nombres de los aeropuertos que contengan la palabra President.
%%sql
select nombre
from aeropuerto
where nombre like '%President%'
```

```python
#La cantidad de aeropuertos que hay Chile.
%%sql
select count(*)
from aeropuerto
where pais='Chile'
```

```python
#La cantidad de ciudades distintas que tienen aeropuertos en Estados Unidos.
%%sql
select count(distinct ciudad)
from aeropuerto
where pais='United States'
```
