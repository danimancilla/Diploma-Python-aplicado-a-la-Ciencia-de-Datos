```python
#P1.La cantidad de vuelos que salen de cada aeropuerto de Chile.
#Consideré que si dos aerloíneas ofrecen el mismo vuelo, entonces son vuelo distintos, por eso no puse un DISTINCT
%%sql
SELECT A.nombre, COUNT(*)
FROM aeropuerto A, ruta R
WHERE A.pais = 'Chile'
AND A.iata=R.origen
GROUP BY A.nombre
```

```python
#P2.Los códigos de los aeropuertos a los que se puede llegar con vuelos desde Santiago (SCL) y no desde Quito (UIO).
%%sql
--Todos los aeropuertos a los que se puede llegar desde SCL
SELECT DISTINCT A.iata
FROM aeropuerto A, ruta R
WHERE R.origen = 'SCL'
AND A.iata=R.destino
--Le quito los aeropuertos a los que se puede llegar desde UIO
EXCEPT
SELECT DISTINCT A.iata
FROM aeropuerto A, ruta R
WHERE R.origen = 'UIO'
AND A.iata=R.destino
```

```python
#P3.Los nombres y ciudades de los aeropuertos que pueden ser visitados desde el aeropuerto de Santiago (SCL).
%%sql
SELECT DISTINCT A.nombre, A.ciudad
FROM aeropuerto A, ruta R
WHERE R.origen = 'SCL'
AND R.destino=A.iata
```

```python
#P4.Las rutas tales que los aeropuertos de origen y el destino sean chilenos.
%%sql
SELECT  R.*
FROM aeropuerto A1, aeropuerto A2, ruta R
WHERE A1.pais = 'Chile'
AND A2.pais = 'Chile'
AND (A1.iata=R.origen AND A2.iata=R.destino)
```

```python
#P5.La cantidad de vuelos que llegan a cada país, ordenados de mayor a menor.
#Consideré que es sin vuelos nacionales, pues dice "llegan" a cada país
%%sql
SELECT A1.pais, COUNT(*)
FROM aeropuerto A1, aeropuerto A2, ruta R
WHERE A1.iata=R.destino
AND A2.iata= R.origen
AND A1.pais <> A2.pais
GROUP BY A1.pais
ORDER BY COUNT(*) DESC
```
