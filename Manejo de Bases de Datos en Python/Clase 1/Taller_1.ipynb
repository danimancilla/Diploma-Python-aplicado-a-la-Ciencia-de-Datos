```python
%load_ext sql
%config SqlMagic.feedback=False
%config SqlMagic.autopandas=True
%sql postgresql+psycopg2://@/postgres
```

```python
#crear tabla aerolinea
%%sql
create table aerolinea(
  iata varchar(2) primary key,
  nombre varchar(500),
  pais varchar(500)
)
```

```python
#crear tabla aeropuerto
#en un comienzo había puesto unique al nombre del aeropuerto,
#pero luego me di cuenta que podían haber dos aeropuertos con el mismo nombre
%%sql
create table aeropuerto(
  iata varchar(3) primary key,
  nombre varchar(500),
  ciudad varchar(500),
  pais varchar(500)
)
```

```python
#crear tabla ruta
%%sql
create table ruta(
  aerolinea varchar(2),
  origen varchar(3),
  destino varchar(3),
  paradas integer CHECK (paradas>=0),
  primary key (aerolinea, origen, destino, paradas)
)
```

```python
#cargar datos desde aerchivos
aerolineas=open("aerolineas.csv")
aeropuertos=open("aeropuertos.csv")
rutas=open("rutas.csv")
```

```python
for linea in aerolineas:
  partes=linea.split(";")
  iata=partes[0]
  nombre=partes[1]
  pais=partes[2].replace("\n","")
  %sql INSERT INTO aerolinea VALUES(:iata, :nombre,:pais)
```

```python
for linea in aeropuertos:
  partes=linea.split(";")
  iata=partes[0]
  nombre=partes[1]
  ciudad=partes[2]
  pais=partes[3].replace("\n","")
  %sql INSERT INTO aeropuerto VALUES(:iata, :nombre,:ciudad, :pais)
```

```python
for linea in rutas:
  partes=linea.split(";")
  aerolinea=partes[0]
  origen=partes[1]
  destino=partes[2]
  paradas=partes[3].replace("\n","")
  %sql INSERT INTO ruta VALUES(:aerolinea, :origen, :destino, :paradas)
```
