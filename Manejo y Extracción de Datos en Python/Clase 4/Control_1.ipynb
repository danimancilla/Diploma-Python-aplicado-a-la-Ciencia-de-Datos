### 3.- Creaci√≥n de Columnas y Rankings üìà

1.- Calcule los ingresos totales que se tienen por fila y almacenelos en una nueva columna llamada `Total`. Comente que variables son las que debe multiplicar para obtener el total por registro.

````python
#Las variables que se deben multiplicar son el precio (Price) y la cantidad (Quantity).
df_copia["Total"]=df_copia["Price"]*df_copia["Quantity"]
df_copia
````

2.- Utilizando un `.groupby()` por la columna `Description`, realice las siguientes agreaciones:

- Suma, m√≠nimo y m√°ximo realizando una agregaci√≥n sobre la columna `Total`.
- Suma, Media, desviaci√≥n estandar, m√≠nimo y m√°ximo realizando la agreaci√≥n sobre la columna `Quantity`.

Guarde los resultados en un nuevo dataframe llamado `df_summary` e imprima los resultados.

````python
df_nuevo=df_copia.loc[:,["Description", "Quantity","Total"]].copy()
df_nuevo
````

````python
df_quantity=df_nuevo.groupby('Description').agg({'Quantity':['sum','mean', 'std', 'min', 'max']}).round(2)
df_quantity
````

````python
df_total=df_nuevo.groupby('Description').agg({'Total':['sum', 'min', 'max']}).round(2)
df_total
````

````python
df_summary = pd.concat([df_total, df_quantity], axis=1)
df_summary
````

3.- Interprete cada una de las columnas obtenidas desde la agregaci√≥n, se√±alando de forma breve que representan los valores obtenidos.

`Para Total, sum indica todo el ingreso asociado al producto (todas las compras, cuanto se recaud√≥), min indica el ingreso m√≠nimo asociado al producto (una compra en particular), y max indica el ingreso m√°ximo asociado al producto (una compra en particular).
Para Quantity, sum indica la cantidad de productos vendidos, asociado a todas las compras.
min indica la m√≠nima cantidad vendida y max indica la m√°xima cantidad vendida. mean es el promedio de productos vendidos en cada compra y std la desviaci√≥n est√°ndar de la cantidad de productos vendidos.`

4.- Utilizando `df_summary` obtenga el top 10 productos m√°s comprados (producto con m√°s n√∫mero de ventas) y el top 10 productos que generan m√°s ingresos (producto que deja m√°s dinero a la tienda) en la tienda de retail. Imprima cada uno de los top se√±alando solo el producto y el dato pertinente al ranking solamente.

- ¬øSon acaso los mismos productos para ambos rankings?

````python
#Top 10 m√°s comprados
df_summary.sort_values(('Quantity',  'sum'), ascending=False).head(10).loc[:,('Quantity',  'sum')]
````

````python
#Top 10 m√°s ingresos
df_summary.sort_values(('Total',  'sum'), ascending=False).head(10).loc[:,('Total',  'sum')]
````

`Algunos productos aparecen en ambos r√°nkings, pero claramente no son los mismos. El top 1 s√≠ coincide.`

4.- En base a `df_summary` genere los quintiles `["Muy bajo", "Bajo", "Medio", "Medio alto", "Alto"]` sobre la columna `('Total', 'sum')`.

````python
df_summary["Quintiles"]=pd.qcut(df_summary[('Total',  'sum')],q=5,labels = ["Muy bajo", "Bajo", "Medio", "Medio alto", "Alto"])
````
