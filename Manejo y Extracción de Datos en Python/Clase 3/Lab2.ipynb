### Limpieza de Datos

1.- Para evitar modificar (y quizas corromper) los datos originales, copie el dataframe en una nueva variable `df_limpio`.

```python
df_limpio=df.copy()
df_limpio
```

2.- Elimine las filas con nombres duplicados.

```python
query=df_limpio.duplicated(subset=['names'])
df_limpio=df_limpio.loc[query==False,:]
df_limpio
```

3.- Elimine todas las filas con valores nulos `NaN`.

```python
df_limpio.isnull().sum()
```

```python
query1=df_limpio["math score"].isnull()
query2=df_limpio["reading score"].isnull()
query3=df_limpio["writing score"].isnull()
query= query1 | query2 | query3
df_limpio=df_limpio.loc[query==False,:]
df_limpio
```

4.- Filtre todas las filas que contengan notas en `math score`, `reading score`, `writing score` mayores a 100.

```python
df_limpio.loc[:,["math score", "reading score", "writing score"]]
query1=df_limpio.loc[:,"math score"]<=100
query2=df_limpio.loc[:,"reading score"]<=100
query3=df_limpio.loc[:,"writing score"]<=100
query= query1 & query2 & query3
df_limpio=df_limpio.loc[query,:]
df_limpio
```

### Tratamiento de Datos

1. Transforme todas las notas porcentuales a la escala chilena, para esto utilice la ecuaci√≥n dispuesta m√°s abajo y aplic√°ndola solamente en las notas porcentuales. El desarrollo de esto deber√° ser sin uso de `for` y aprovechando los beneficios de pandas.

$$ Nota \, Chilena= \% \, de \, logro*0.06 + 1$$

```python
nota_math=df_limpio["math score"]*0.06+1
nota_reading=df_limpio["reading score"]*0.06+1
nota_writing=df_limpio["writing score"]*0.06+1
```

2. Para cada uno de los alumnos cree una nueva columna que posea el promedio de todas las asignaturas cursadas. Tenga precauci√≥n en que las notas promediadas est√©n en la escala de notas chilena.

```python
promedio=(nota_math+nota_reading+nota_writing)/3
df_limpio["Promedio"]=promedio
df_limpio
```

3.- Utilizando el promedio general calculado en la pregunta anterior, identifique los alumnos que reprobaron el semestre (promedio inferior a 4.0).

```python
query=df_limpio.loc[:,"Promedio"]<4.0
df_limpio.loc[query,:]
```
