### 1.- Creaci칩n de Columnas y Rankings 游늳

Cargue el archivo [`online_retail_II.pickle`](https://raw.githubusercontent.com/pbadillatorrealba/DEDCCPYAC02-recursos/master/labs/lab3/online_retail_II.pickle) utilizando pandas.

````python
df=pd.read_pickle("online_retail_II.pickle")
````

````python
df.head(5)
````

1.- Calcule los ingresos totales que se tienen por fila y almacenelos en una nueva columna llamada `Total`. Comente que variables son las que debe multiplicar para obtener el total por registro.

````python
df.Quantity*df.Price
df["Total"]=df.Quantity*df.Price
````

Se debe multiplicar la cantidad por el precio.

2.- Utilizando un `.groupby()` por la columna `Description`, realice las siguientes agreaciones:

- Suma, m칤nimo y m치ximo realizando una agregaci칩n sobre la columna `Total`.
- Suma, Media, desviaci칩n estandar, m칤nimo y m치ximo realizando la agreaci칩n sobre la columna `Quantity`.

Guarde los resultados en un nuevo dataframe llamado `df_summary` e imprima los resultados.

Primero borrar칠 los valores nulos.

````python
df_copia=df.copy()
no_nulos=~df_copia.isna().any(axis=1)
df_copia=df_copia.loc[no_nulos]
df_copia.head(5)
````

````python
df_summary=df_copia.groupby(["Description"]).agg({"Total": ["sum", "min", "max"], "Quantity":["sum", "mean", "std", "min" , "max"]}).round(2)
````

````python
df_summary.head(10)
````

3.- Interprete cada una de las columnas obtenidas desde la agregaci칩n, se침alando de forma breve que representan los valores obtenidos.

Para Total, sum indica todo el ingreso asociado al producto (considera todas las compras, cuanto se recaud칩 en total), min indica el ingreso m칤nimo asociado al producto (una compra en particular), y max indica el ingreso m치ximo asociado al producto (una compra en particular).
Para Quantity, sum indica la cantidad total que se vendi칩 para cada producto, dasociado a todas las compras. min indica la m칤nima cantidad vendida y max indica la m치xima cantidad vendida, ambas en una compra en particular. mean es el promedio de productos vendidos por compra y std la desviaci칩n est치ndar de la cantidad de productos vendidos.
Cuando aparecen valores negativos es porque las personas "devolvieron" los productos que hab칤an escaneado.

4.- Utilizando `df_summary` obtenga el top 10 productos m치s comprados (producto con m치s n칰mero de ventas) y el top 10 productos que generan m치s ingresos (producto que deja m치s dinero a la tienda) en la tienda de retail. Imprima cada uno de los top se침alando solo el producto y el dato pertinente al ranking solamente.

- 쯉on acaso los mismos productos para ambos rankings?

````python
#Top 10 m치s comprados
df_summary.sort_values(('Quantity',  'sum'), ascending=False).head(10).loc[:,('Quantity',  'sum')]
````

````python
#Top 10 m치s ingresos
df_summary.sort_values(('Total',  'sum'), ascending=False).head(10).loc[:,('Total',  'sum')]
````

Algunos productos aparecen en ambos r치nkings, pero claramente no son los mismos. El top 1 s칤 coincide.
La diferencia en los r치nkings viene de que los productos tienen distintos precios, entonces aunque algunos productos sean muy vendidos, si sus precios no son tan altos, los ingresos tampoco lo ser치n.

5.- En base a `df_summary` genere los quintiles `["Muy bajo", "Bajo", "Medio", "Medio alto", "Alto"]` sobre la columna `('Total', 'sum')`.

````python
df_summary["Quintiles"]=pd.qcut(df_summary[('Total',  'sum')],q=5,labels = ["Muy bajo", "Bajo", "Medio", "Medio alto", "Alto"])
df_summary
````
