### 1.- CreaciÃ³n de Columnas y Rankings ðŸ“ˆ

Cargue el archivo [`online_retail_II.pickle`](https://raw.githubusercontent.com/pbadillatorrealba/DEDCCPYAC02-recursos/master/labs/lab3/online_retail_II.pickle) utilizando pandas.

````python
df=pd.read_pickle("online_retail_II.pickle")
````

````python
df.head(5)
````

1.- Calcule los ingresos totales que se tienen por fila y almacenelos en una nueva columna llamada `Total`. Comente que variables son las que debe multiplicar para obtener el total por registro.

````python
df.Quantity*df.Price
df["Total"]=df.Quantity*df.Price
````

Se debe multiplicar la cantidad por el precio.

2.- Utilizando un `.groupby()` por la columna `Description`, realice las siguientes agreaciones:

- Suma, mÃ­nimo y mÃ¡ximo realizando una agregaciÃ³n sobre la columna `Total`.
- Suma, Media, desviaciÃ³n estandar, mÃ­nimo y mÃ¡ximo realizando la agreaciÃ³n sobre la columna `Quantity`.

Guarde los resultados en un nuevo dataframe llamado `df_summary` e imprima los resultados.

Primero borrarÃ© los valores nulos.

````python
df_copia=df.copy()
no_nulos=~df_copia.isna().any(axis=1)
df_copia=df_copia.loc[no_nulos]
df_copia.head(5)
````

````python
df_summary=df_copia.groupby(["Description"]).agg({"Total": ["sum", "min", "max"], "Quantity":["sum", "mean", "std", "min" , "max"]}).round(2)
````

````python
df_summary.head(10)
````

3.- Interprete cada una de las columnas obtenidas desde la agregaciÃ³n, seÃ±alando de forma breve que representan los valores obtenidos.

Para Total, sum indica todo el ingreso asociado al producto (considera todas las compras, cuanto se recaudÃ³ en total), min indica el ingreso mÃ­nimo asociado al producto (una compra en particular), y max indica el ingreso mÃ¡ximo asociado al producto (una compra en particular).
Para Quantity, sum indica la cantidad total que se vendiÃ³ para cada producto, dasociado a todas las compras. min indica la mÃ­nima cantidad vendida y max indica la mÃ¡xima cantidad vendida, ambas en una compra en particular. mean es el promedio de productos vendidos por compra y std la desviaciÃ³n estÃ¡ndar de la cantidad de productos vendidos.
Cuando aparecen valores negativos es porque las personas "devolvieron" los productos que habÃ­an escaneado.

4.- Utilizando `df_summary` obtenga el top 10 productos mÃ¡s comprados (producto con mÃ¡s nÃºmero de ventas) y el top 10 productos que generan mÃ¡s ingresos (producto que deja mÃ¡s dinero a la tienda) en la tienda de retail. Imprima cada uno de los top seÃ±alando solo el producto y el dato pertinente al ranking solamente.

- Â¿Son acaso los mismos productos para ambos rankings?

````python
#Top 10 mÃ¡s comprados
df_summary.sort_values(('Quantity',  'sum'), ascending=False).head(10).loc[:,('Quantity',  'sum')]
````

````python
#Top 10 mÃ¡s ingresos
df_summary.sort_values(('Total',  'sum'), ascending=False).head(10).loc[:,('Total',  'sum')]
````

Algunos productos aparecen en ambos rÃ¡nkings, pero claramente no son los mismos. El top 1 sÃ­ coincide.
La diferencia en los rÃ¡nkings viene de que los productos tienen distintos precios, entonces aunque algunos productos sean muy vendidos, si sus precios no son tan altos, los ingresos tampoco lo serÃ¡n.

5.- En base a `df_summary` genere los quintiles `["Muy bajo", "Bajo", "Medio", "Medio alto", "Alto"]` sobre la columna `('Total', 'sum')`.

````python
df_summary["Quintiles"]=pd.qcut(df_summary[('Total',  'sum')],q=5,labels = ["Muy bajo", "Bajo", "Medio", "Medio alto", "Alto"])
df_summary
````
