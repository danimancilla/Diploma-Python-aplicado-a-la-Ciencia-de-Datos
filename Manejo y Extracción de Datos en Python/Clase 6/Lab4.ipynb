````python
df_other =pd.read_csv("https://raw.githubusercontent.com/pbadillatorrealba/DEDCCPYAC02-recursos/master/labs/lab4/other_grades.csv")
df_other
````

````python
#Inner
pd.merge(left=df_grades,
         right=df_other,
         on="names",
         how="inner",
         sort=True)
````

```
Inner merge busca los elementos comunes de las columnas elegidas,
en este caso "names", y combina la informaci√≥n de ambos dataframes en uno solo.
En este caso, encontr√≥ 875 estudiantes comunes, y agreg√≥ las 3 columnas del dataframe 2 que no eran "names",
a la derecha del primer dataframe.
```

````python
#Outer
pd.merge(left=df_grades,
         right=df_other,
         on="names",
         how="outer",
         sort=True)
````

```
Outer merge combina todos los elementos de ambos dataframes. En este caso, hicimos el merge con la columna "names",
si un nombre no est√° en un dataframe,
rellena con NaN la informaci√≥n que deber√≠a estar guardada ah√≠.


```

````python
#Right
pd.merge(left=df_grades,
         right=df_other,
         on="names",
         how="right",
         sort=True)
````

```
Right merge mantiene todos los elementos del dataframe elegido como right, que en este caso son 1000.
Agrega a la izquierda los elementos del dataframe left, pero solo los que comparta con el de la derecha.
Si un nombre de la derecha no est√° en el dataframe de la izquierda, rellena la informaci√≥n con NaN.
```

````python
#Left
pd.merge(left=df_grades,
         right=df_other,
         on="names",
         how="left",
         sort=True)
````

```
Left merge mantiene todos los elementos del dataframe elegido como left, que en este caso son 875.
Agrega a la derecha los elementos del dataframe right, pero solo los que comparta con el de la izquierda.
Si un nombre de la izquierda no est√° en el dataframe de la derecha, rellena la informaci√≥n con NaN.```

````python
df_merged=pd.merge(left=df_grades,
         right=df_other,
         on="names",
         how="inner",
         sort=True).drop('Unnamed: 0', axis=1)
df_merged
````

3. En los siguientes tres puntos deber√° generar un detector de sentimientos basado en la cantidad de palabras positivas y negativas que posee cada review. El primer paso para esto es:

- Preprocesar la columna `review` convirtiendo todos los caracteres a min√∫sculas y elimine todos los acentos y puntuaciones. Para eliminar puntuaciones, efectue una b√∫squeda en google en ingles del estilo *como eliminar puntuaciones en pandas*.
- Separar las palabras de cada fila usando espacio como separador.

````python
reviews=df_reviews["review"].str.lower().str.replace('[^\w\s]', " ",regex=True).str.replace('\s+', ' ', regex=True).str.split(" ")
reviews
````

4.- Genere las series `positive` y ``negative`` aplicando sobre las `reviews` procesadas las funciones `contar_positivas` y `contar_negativas`. As√≠gnelos resultados obtenidos a una nueva columna.

````python
# El siguiente c√≥digo contiene las palabras positivas y negativas que ser√°n contadas en cada review.
# No deber√≠an modificarlo, solo aplicarlo para generar las series que cuentan cuantos positivos y negativos tiene cada fila.

palabras_positivas = [
    "acclaimed",
    "affirmative",
    "affluent",
    "angelic",
    "beaming",
    "bliss",
    "bountiful",
    "bubbly",
    "commend",
    "dazzling",
    "earnest",
    "ecstatic",
    "effervescent",
    "enchanting",
    "exquisite",
    "graceful",
    "instinctive",
    "intellectual",
    "jovial",
    "jubilant",
    "lucid",
    "marvelous",
    "novel",
    "nurturing",
    "poised",
    "robust",
    "stirring",
    "stupendous",
    "thriving",
    "tranquil",
    "unwavering",
    "upbeat",
    "virtuous",
    "vivacious",
    "zealous",
]


def contar_positivas(lista_palabras):
    numero_palabras_positivas = 0
    for palabra_positiva in palabras_positivas:
        if palabra_positiva in lista_palabras:
            numero_palabras_positivas += 1

    return numero_palabras_positivas


palabras_negativas = [
    "abysmal",
    "apathy",
    "atrocious",
    "bemoan",
    "cold-hearted",
    "deplorable",
    "despicable",
    "detrimental",
    "dreadful",
    "ghastly",
    "grotesque",
    "gruesome",
    "hard-hearted",
    "insidious",
    "insipid",
    "malicious",
    "monstrous",
    "noxious",
    "offensive",
    "oppressive",
    "pessimistic",
    "prejudice",
    "repulsive",
    "ruthless",
    "sinister",
    "vice",
    "vicious",
    "vile",
    "villainous",
    "vindictive",
]


def contar_negativas(lista_palabras):
    numero_palabras_negativas = 0
    for palabra_positiva in palabras_negativas:
        if palabra_positiva in lista_palabras:
            numero_palabras_negativas += 1

    return numero_palabras_negativas
````

````python
reviews.apply(contar_positivas)
````

````python
reviews.apply(contar_negativas)
````

````python
df_reviews["Positive"]=reviews.apply(contar_positivas)
df_reviews["Negative"]=reviews.apply(contar_negativas)
df_reviews
````

5.-Implemente la funci√≥n `detectar_sentimientos` que toma como entrada una fila del dataframe y retorna el sentimiento de cada oraci√≥n. La condici√≥n que debe verificar es que:

- $positiva - negativa = 0$, entonces retorna `undefined`.
- $positiva > negativa$, entonces retorna 'positive'.
- $positiva < negativa$, entonces retorna `negative`.

Aplique la funci√≥n detectar_sentimientos sobre las filas del dataset para calcular el sentimiento de cada review. Luego, asigne la serie generada al dataframe original.

````python
def detectar_sentimientos(fila):
  positiva=fila["Positive"]
  negativa=fila["Negative"]
  if positiva-negativa==0:
    return "undefined"
  elif positiva > negativa:
    return "positive"
  else:
    return "negative"
````

````python
sentimiento=df_reviews.apply(detectar_sentimientos,axis=1)
sentimiento
````

````python
df_reviews["Sentimiento"]=sentimiento
df_reviews
````
