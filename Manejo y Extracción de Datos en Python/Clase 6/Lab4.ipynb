4.- Genere las series `positive` y ``negative`` aplicando sobre las `reviews` procesadas las funciones `contar_positivas` y `contar_negativas`. As铆gnelos resultados obtenidos a una nueva columna.

````python
# El siguiente c贸digo contiene las palabras positivas y negativas que ser谩n contadas en cada review.
# No deber铆an modificarlo, solo aplicarlo para generar las series que cuentan cuantos positivos y negativos tiene cada fila.

palabras_positivas = [
    "acclaimed",
    "affirmative",
    "affluent",
    "angelic",
    "beaming",
    "bliss",
    "bountiful",
    "bubbly",
    "commend",
    "dazzling",
    "earnest",
    "ecstatic",
    "effervescent",
    "enchanting",
    "exquisite",
    "graceful",
    "instinctive",
    "intellectual",
    "jovial",
    "jubilant",
    "lucid",
    "marvelous",
    "novel",
    "nurturing",
    "poised",
    "robust",
    "stirring",
    "stupendous",
    "thriving",
    "tranquil",
    "unwavering",
    "upbeat",
    "virtuous",
    "vivacious",
    "zealous",
]


def contar_positivas(lista_palabras):
    numero_palabras_positivas = 0
    for palabra_positiva in palabras_positivas:
        if palabra_positiva in lista_palabras:
            numero_palabras_positivas += 1

    return numero_palabras_positivas


palabras_negativas = [
    "abysmal",
    "apathy",
    "atrocious",
    "bemoan",
    "cold-hearted",
    "deplorable",
    "despicable",
    "detrimental",
    "dreadful",
    "ghastly",
    "grotesque",
    "gruesome",
    "hard-hearted",
    "insidious",
    "insipid",
    "malicious",
    "monstrous",
    "noxious",
    "offensive",
    "oppressive",
    "pessimistic",
    "prejudice",
    "repulsive",
    "ruthless",
    "sinister",
    "vice",
    "vicious",
    "vile",
    "villainous",
    "vindictive",
]


def contar_negativas(lista_palabras):
    numero_palabras_negativas = 0
    for palabra_positiva in palabras_negativas:
        if palabra_positiva in lista_palabras:
            numero_palabras_negativas += 1

    return numero_palabras_negativas
````

````python
reviews.apply(contar_positivas)
````

````python
reviews.apply(contar_negativas)
````

````python
df_reviews["Positive"]=reviews.apply(contar_positivas)
df_reviews["Negative"]=reviews.apply(contar_negativas)
df_reviews
````

5.-Implemente la funci贸n `detectar_sentimientos` que toma como entrada una fila del dataframe y retorna el sentimiento de cada oraci贸n. La condici贸n que debe verificar es que:

- $positiva - negativa = 0$, entonces retorna `undefined`.
- $positiva > negativa$, entonces retorna 'positive'.
- $positiva < negativa$, entonces retorna `negative`.

Aplique la funci贸n detectar_sentimientos sobre las filas del dataset para calcular el sentimiento de cada review. Luego, asigne la serie generada al dataframe original.

````python
def detectar_sentimientos(fila):
  positiva=fila["Positive"]
  negativa=fila["Negative"]
  if positiva-negativa==0:
    return "undefined"
  elif positiva > negativa:
    return "positive"
  else:
    return "negative"
````

````python
sentimiento=df_reviews.apply(detectar_sentimientos,axis=1)
sentimiento
````

````python
df_reviews["Sentimiento"]=sentimiento
df_reviews
````
